```python
import pandas as pd
```

#### Data collection
The following cell will read in the selected dataset containing the us wind turbine info, which we will then read into a pandas dataframe for data manipulation

```python
df = pd.read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2018/2018-11-06/us_wind.csv")
```

```python
df.sample(7)
```

#### Data Filteration

From the above cell which shows us a small sample from the dataset, and upon cross checking with the information given from the github page, we've chosen to filter out the following colums: 

```python
df2=df.drop(['case_id',
             'faa_ors',
             'faa_asn',
             'usgs_pr_id',
             't_county',
             't_fips',
             't_model',
             't_hh',
             't_rd',
             't_rsa',
             't_ttlh',
             't_conf_atr',
             't_conf_loc',
             't_img_date',
             't_img_srce'], axis=1)
```

```python
df2.sample(7)
```

Now we will rename the columns' names to allow for easier usage in our vega lite file

```python
df2 = df2.rename(columns={"t_state": "turbineState",
                          "p_name" : "projectName",
                          "p_year" : "projectYear",
                          "p_tnum" : "projectTurbineNumber",
                          "p_cap" : "projectCapacity",
                          "t_manu" : "turbineManufacturer",
                          "t_cap" : "turbineCapacity",
                          "xlong" : "long",
                          "ylat" : "lat"})
```

```python
df2.sample(7)
```

Next any missing/unknown data will be filtered out, as well as any data with negligible data such as wind turbines in Purerto Rico

```python
df2 = df2[df2['projectCapacity'] != -9999]
df2 = df2[df2['projectCapacity'] >= 0]
df2 = df2[df2['projectYear'] != -9999]
df2 = df2[df2['projectTurbineNumber'] != -9999]
df2 = df2[df2['projectTurbineNumber'] >= 0]
df2 = df2[df2["turbineManufacturer"].str.contains("missing") == False]
df2 = df2[df2["turbineState"].str.contains("PR") == False]
df2 = df2[df2["projectName"].str.contains("unknown") == False]
df2 = df2[df2["projectName"].str.contains("missing") == False]
```

```python
df2.sample(7)
```

```python
df2 = df2.groupby('turbineManufacturer').filter(lambda x : len(x)>100)
```

```python
df2["turbineCapacity"].describe()
```

Next we will add another column which will aid our vega lite map graph to allow selection of turbines of certain power capacities

```python
df2['TurbineCapacityCat'] = pd.cut(df2["turbineCapacity"],
                                   bins=[0, 1500, 2050, float('Inf')],
                                   labels=['Small', 'Medium', 'Large'])
```

```python
df2['TurbineCapacityCat'].value_counts()
```

```python
df2 = df2.groupby('turbineManufacturer').filter(lambda x : len(x)>100)
```

The next step is to replace all state name abbrevations to their full state names

```python
states = {
            'AK': 'Alaska',
            'AL': 'Alabama',
            'AR': 'Arkansas',
            'AS': 'American Samoa',
            'AZ': 'Arizona',
            'CA': 'California',
            'CO': 'Colorado',
            'CT': 'Connecticut',
            'DC': 'District of Columbia',
            'DE': 'Delaware',
            'FL': 'Florida',
            'GA': 'Georgia',
            'GU': 'Guam',
            'HI': 'Hawaii',
            'IA': 'Iowa',
            'ID': 'Idaho',
            'IL': 'Illinois',
            'IN': 'Indiana',
            'KS': 'Kansas',
            'KY': 'Kentucky',
            'LA': 'Louisiana',
            'MA': 'Massachusetts',
            'MD': 'Maryland',
            'ME': 'Maine',
            'MI': 'Michigan',
            'MN': 'Minnesota',
            'MO': 'Missouri',
            'MP': 'Northern Mariana Islands',
            'MS': 'Mississippi',
            'MT': 'Montana',
            'NA': 'National',
            'NC': 'North Carolina',
            'ND': 'North Dakota',
            'NE': 'Nebraska',
            'NH': 'New Hampshire',
            'NJ': 'New Jersey',
            'NM': 'New Mexico',
            'NV': 'Nevada',
            'NY': 'New York',
            'OH': 'Ohio',
            'OK': 'Oklahoma',
            'OR': 'Oregon',
            'PA': 'Pennsylvania',
            'PR': 'Puerto Rico',
            'RI': 'Rhode Island',
            'SC': 'South Carolina',
            'SD': 'South Dakota',
            'TN': 'Tennessee',
            'TX': 'Texas',
            'UT': 'Utah',
            'VA': 'Virginia',
            'VI': 'Virgin Islands',
            'VT': 'Vermont',
            'WA': 'Washington',
            'WI': 'Wisconsin',
            'WV': 'West Virginia',
            'WY': 'Wyoming'
    }

state2 = {
    'Siemens Gamesa Renewable Energy' : 'SGRE'
}

df2['turbineState'] = df2['turbineState'].replace(states)
df2['turbineManufacturer'] = df2['turbineManufacturer'].replace(state2)
```
